```python
!pip install jamo

from jamo import h2j, j2hcj
def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python
def make_user_dic_csv(morpheme_type, word_list, user_dic_file_name):
  file_data = []

  for word, score in word_list:
    jongsung_TF = get_jongsung_TF(word)

    line = f"{word},,,{score},{morpheme_type},*,{jongsung_TF},{word},*,*,*,*,*\n"

    file_data.append(line)

  with open("./user-dic/user-nnp.csv", 'w', encoding='utf-8') as f:
    for line in file_data:
      f.write(line)
```

```python
word_list = [('펭수', 0)]

make_user_dic_csv(morpheme_type="NNP", word_list=word_list, user_dic_file_name='user-nnp.csv')
```

## LDA 토픽 모델링 분석
LDA를 활용하여 토픽을 추출합니다. 토픽 모델링은 하이퍼파라미터 튜닝과 전처리가 중요합니다.</br>
그리고 나온 결괏값에 대한 해석은 사용자의 몫입니다.</br>
끝으로 문서별 토픽을 확인합니다.

```python
# 토큰화된 문서
tokenized_doc = df['본문_전처리']
```

```python
# Gensim에서 지원하는 LDA 모델 불러오기
from gensim.models.ldamodel import LdaModel
from gensim import corpora

dictionary = corpora.Dictionary(tokenized_doc)
```

```python
# 단어 말뭉치(Corpus) 생성
corpus = [dictionary.doc2bow(text) for text in tokenized_doc] # document를 Bag of Words로
corpus
```

```python
NUM_TOPICS = 10 # 토픽 개수 지정 (하이퍼 파라미터)
ldamodel = LdaModel(corpus,
                    num_topics = NUM_TOPICS,
                    id2word = dictionary,
                    passes = 30) # 학습 횟수

topics = ldamodel.print_topics(num_words = 10) # 토픽 저장
```

```python
topics
```

```python
import pyLDAvis.gensim_models

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```
