```python
# '삼성'이 들어간 제목만 필터링

# 신약, 주가 '주가' '증권' '배당' '순이익' '주식' '매수' '매도'
df = df[df['단어'].map(lambda x: '삼성' in x)].copy();df

df = df[df['단어'].map(lambda x: '주가' in x or '증권' in x or '배당' in x or '주식' in x or '매수' in x  or '매도' in x)].copy(); df
```

```python
# 자연어 전처리: 텍스트 클리닝
import re
df['제목_전처리'] = df['제목'].map(lambda x: re.sub(r'[^\s\w]', ' ', x))
```

```python
df['제목_전처리']
```

```python
# 인덱스 리셋
df.reset_index(drop=True, inplace=True)
```

```python
# TF-IDF 벡터화
vectorizer = TfidfVectorizer(min_df=10, ngram_range=(1,2), tokenizer = okt.morphs) #okt.nouns
features = vectorizer.fit_transform(df['제목_전처리'])
```

```python
# feature 이름 불러 오기
feature_names = vectorizer.get_feature_names_out()
feature_names
```

```python
# DTM 생성
dtm_np = np.array(features.todense()); dtm_np
```

```python
# DataFrame 확인
pd.DataFrame(data = dtm_np, columns = feature_names)
```

```python
# 코사인 유사도 구하기
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(dtm_np, dtm_np) #a 행렬과 b 행렬, 자기 자신과의 코사인 유사도 구하기 ...
```

```python
# 인덱스 설정
indices = pd.Series(df.index, index=df['제목']); indices
```

```python
# 뉴스 기사 제목 확인 하기
set(df['제목'])
```

```python
# 뉴스 기사 인덱스 추출
idx = indices['“삼성전자 주식, 개인이 왜 사요?” 24년차 IT애널의 투자 조언[딥다이브]']; print(idx)
```

```python
# 선택한 뉴스 기사와 다른 뉴스 기사간 유사도 확인
sim_scores = list(enumerate(cosine_sim[idx]))
```

```python
# 유사도가 높은 순으로 뉴스 정렬
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
```

```python
# 유사도 높은 10개의 뉴스
sim_scores = sim_scores[1:11] # 0은 자기 자신일 것

# 가장 유사한 10개의 뉴스의 인덱스
news_indices = [i[0] for i in sim_scores]

# 유사한 뉴스 기사 제목 출력
df['제목'].iloc[news_indices]

# 코사인 유사도 기반 데이터 프레임
sim_df = pd.DataFrame(cosine_sim); sim_df
```

```python
# Boolean Index 만들기
sim_boolean = sim_df > 0.5; sim_boolean
```

```python
# 유사도 0.5가 넘는 게시글이 10개 이상 있는 경우
sim_boolean.sum() >= 10
```

```python
# 10개 이상 있는 게시글이 몇 개나 있을까?
sum(sim_boolean.sum() >= 10)
```

```python
# 유사도가 0.5 이상인 뉴스의 인덱스 확인해보기
sim_df[sim_boolean.sum() >= 10].index
```

```python
# 유사도가 0.4 이상인 게시글은 얼마나 유사할까?
idx_list = sim_df[0][sim_df[0] > 0.4].index
df.iloc[idx_list,:]
```

```python
# 특정 유사도 이상의 뉴스 토픽 묶기

idx_list = list(sim_df[sim_boolean.sum() >= 10].index)
cluster = []
pass_list = []
threshold = 0.25
id_idx = []
for i in idx_list:
    if i not in pass_list:
        idx = sim_df[i][sim_df[i] > threshold].index
        cluster.append(idx)
        pass_list.extend(list(idx))
        id_idx.append(i)
    else:
        pass

print(len(cluster))
```

```python
# 삼성 주식 뉴스 토픽
samsung_df = df.iloc[id_idx,:].copy(); samsung_df
```
